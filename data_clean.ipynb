```python
import os
import hashlib
import pandas as pd

input_folder = r"LinkedIn Data Public"
output_folder = "cleaned"
os.makedirs(output_folder, exist_ok=True)

def get_file_hash(filepath):
    with open(filepath, "rb") as f:
        return hashlib.md5(f.read()).hexdigest()

def format_filename(raw_name):
    name_part = os.path.splitext(raw_name)[0]
    if " - " in name_part:
        name_part = name_part.split(" - ", 1)[1]
    name_part = name_part.replace("_", " ").replace("-", " ")
    words = [w.capitalize() for w in name_part.split()]
    return " ".join(words).strip() or "Unnamed"

def try_read_csv(filepath):
    encodings = ['utf-8', 'ISO-8859-1', 'cp1252']
    for enc in encodings:
        try:
            return pd.read_csv(filepath, encoding=enc, on_bad_lines='skip', engine='python')
        except Exception:
            continue
    return pd.read_csv(filepath, encoding='utf-8', errors='replace', on_bad_lines='skip', engine='python')

seen_hashes = set()
errors = []

for filename in os.listdir(input_folder):
    if not filename.lower().endswith(".csv"):
        continue

    file_path = os.path.join(input_folder, filename)
    file_hash = get_file_hash(file_path)

    if file_hash in seen_hashes:
        continue
    seen_hashes.add(file_hash)

    try:
        df = try_read_csv(file_path)
        df = df.drop_duplicates()
        df.columns = [col.strip() for col in df.columns]
        df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

        clean_name = format_filename(filename)
        output_path = os.path.join(output_folder, f"{clean_name}.csv")
        df.to_csv(output_path, index=False)
        print(f"âœ… Cleaned: {filename} â†’ {clean_name}.csv")
    except Exception as e:
        errors.append((filename, str(e)))
        print(f"âŒ Error in '{filename}': {e}")

print(f"\nðŸŽ‰ All Done! Cleaned files saved to: {output_folder}")
if errors:
    print("\nâš ï¸ Still had issues with:")
    for fname, err in errors:
        print(f" - {fname}: {err}")
```

```python
import os
cleaned_path = "cleaned"
public_path = "LinkedIn Data Public"

def normalize_name(name):
    """Normalize name: lowercase, remove file extension, hyphens, underscores, extra spaces"""
    name = os.path.splitext(name)[0]
    name = name.replace('_', ' ').replace('-', ' ').lower()
    return ' '.join(name.split())  

def extract_cleaned_names(path):
    """Get normalized names from the cleaned folder"""
    return {normalize_name(f) for f in os.listdir(path)}

def extract_public_names(path):
    """Extract names after hyphen in the public folder and normalize them"""
    name_map = {}
    for f in os.listdir(path):
        base = os.path.splitext(f)[0]
        if '-' in base:
            name_part = base.split('-')[1].strip()
        else:
            name_part = base.strip()
        normalized = normalize_name(name_part)
        name_map[normalized] = f
    return name_map
cleaned_names = extract_cleaned_names(cleaned_path)
public_name_map = extract_public_names(public_path)

missing_in_cleaned = {name: public_name_map[name] for name in public_name_map if name not in cleaned_names}


print("ðŸ“‚ Summary Report")
print("----------------------------")
print(f"âœ… Total files in LinkedIn Data Public: {len(public_name_map)}")
print(f"âœ… Total files in Cleaned: {len(cleaned_names)}")
print(f"âŒ Files from Public NOT in Cleaned: {len(missing_in_cleaned)}")
print()

# Show missing file names
if missing_in_cleaned:
    print("ðŸš« Missing Files from Public folder:")
    for norm_name, original_file in missing_in_cleaned.items():
        print(f" - {original_file}")
else:
    print("ðŸŽ‰ All public files are matched in cleaned folder.")
```

```python
import os

public_folder = "LinkedIn Data Public"
cleaned_folder = "cleaned"
def normalize_public_name(filename):
    name = os.path.splitext(filename)[0]
    if '-' in name:
        name = name.split('-')[1].strip()
    return name.replace('_', ' ').replace('-', ' ').strip().lower()

def normalize_cleaned_name(filename):
    name = os.path.splitext(filename)[0]
    return name.strip().lower()
public_files = {
    normalize_public_name(f)
    for f in os.listdir(public_folder)
    if os.path.isfile(os.path.join(public_folder, f))
}

cleaned_files = {
    normalize_cleaned_name(f)
    for f in os.listdir(cleaned_folder)
    if os.path.isfile(os.path.join(cleaned_folder, f))
}

missing_in_cleaned = public_files - cleaned_files
extra_in_cleaned = cleaned_files - public_files

print(f"ðŸ“‚ Files in LinkedIn Data Public: {len(public_files)}")
print(f"ðŸ“‚ Files in cleaned: {len(cleaned_files)}\n")

print(f"âŒ Missing in cleaned ({len(missing_in_cleaned)}):")
for name in sorted(missing_in_cleaned):
    print(" -", name)

print(f"\nâš ï¸ Extra in cleaned but not in public ({len(extra_in_cleaned)}):")
for name in sorted(extra_in_cleaned):
    print(" -", name)
```
